# Напишите программирование на Python для отображения гистограммы популярности языков программирования. Используйте равномерный цвет.  
Пример данных:
Языки программирования: Java, Python, PHP, JavaScript, C #, C ++
Популярность: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [7.7, 6.7, 22.2, 17.6, 8.8]
ax.bar(langs, students, color = 'm')
ax.set_title('Popularity of Programming Language Worldwide, Oct 2017 compared to a year ago')
ax.set_xlabel('Languages')
ax.set_ylabel('Popularity')
plt.show()
```

```
from matplotlib import pyplot as plt
import numpy as np
fig, ax = plt.subplots()
explode = (0.1, 0, 0.15, 0, 0, 0)
labels = ['Java', 'C #', 'C++', 'JavaScipt', 'Python', 'PHP']
students = [22.2, 7.7, 6.7, 8, 17.6, 8.8]
ax.pie(students, labels = labels, autopct='%1.1f%%', shadow = True, explode = explode)
ax.set_title('Popularity of Programming Language Worldwide, Oct 2017 compared to a year ago')
ax.axis('equal')
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("SOCR-HeightWeight.csv", sep=',' , parse_dates = True, index_col = 0)
sum_m = df["Height(Inches)"].sum()
print(sum_m)
sum_f = df["Weight(Pounds)"].sum()
print(sum_f)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
genders = ['Height', 'Weight']
hweight = [sum_m, sum_f]
ax.pie(hweight, labels = genders, colors = ['green', 'teal'], autopct = '%1.2f%%', wedgeprops={'lw':1,'edgecolor':"k"})
ax.set_title("Круговая диаграмма веса и роста")
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("SOCR-HeightWeight.csv", sep = ',', parse_dates=True)
min_m = df["Height(Inches)"].min()
print(min_m)
min_f = df["Weight(Pounds)"].min()
print(min_f)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
hweight = [min_m, min_f]
genders = ['Height', 'Weight']
ax.bar(genders, hweight, color = 'y')
ax.set_title("Гистограммы веса и роста")
ax.set_ylabel('Индекс веса и роста')
plt.show()
```
