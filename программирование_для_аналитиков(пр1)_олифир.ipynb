```
import timeit
import numpy as np 

def vector1(n, x):
  n = n
  x = x
  z = np.array([0 if i != x else 1 for i in range(n)], dtype=float)

def vector2(n, x):
    n = n
    x = x
    z = np.zeros(n)
    z[x] = 1


def nonumpy(n, x):
  n = n
  x = x
  z = [0]*n
  z[x] = 1

if __name__ == '__main__':

  start_time = timeit.default_timer()
  vector1(n, x)
  time_1 = timeit.default_timer() - start_time

  start_time = timeit.default_timer()
  vector2(n, x)
  time_2 = timeit.default_timer() - start_time

  start_time = timeit.default_timer()
  nonumpy(n, x)
  time_3 = timeit.default_timer() - start_time

```

```
print(f'время {time_1}') 
print(f'время {time_2}') 
print(f'время {time_3}') 
```
