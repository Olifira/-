```
class Motor:
  def __init__(self, w, p, s):
    self.weight = w
    self.power = p
    self.size = s

class ExplosionEngine(Motor):
  def expltime(self, m):
    self.moment = m
    return self.moment

class ReactiveEngine(Motor):
  def reactfast(self, f):
    self.fast = f
    return self.fast

class Diesel(ExplosionEngine):
  def diesel(self, t):
    self.temperature = t
    return self.temperature

t1 = ExplosionEngine(130, 170, "70 x 80 mm")
print("weight: {}, power: {}, size: {}, moment: {}". format(t1.weight, t1.power, t1.size, t1.expltime("40 min")))
t2 = ReactiveEngine(100, 200, "90 x 100 mm")
print("weight: {}, power: {}, size: {}, fast: {}". format(t2.weight, t2.power, t2.size, t2.reactfast("30 m/min")))
t3 = Diesel(110, 115, "60 x 70 mm")
print("weight: {}, power: {}, size: {}, temperature: {}". format(t3.weight, t3.power, t3.size, t3.diesel("100C")))
 
```
