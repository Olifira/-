с точным решением. Нарисуйте графики точного и численного решений, а также относительной ошибки численного решения.

y_nZHl9Fl0k.jpg

```
from scipy import integrate 
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(yf, t):
  v = yf[0]
  y = yf[1]
  f0 = -y+np.sin(t)-np.cos(2*t)
  f1 = v
  return[f0, f1]
```

```
ti = np.linspace(0, 7*np.pi, 1000)
x0 = np.pi
v0 = 2.0
y0 = [x0, v0]
```

```
sol = integrate.odeint(f, y0, ti)
y1 = sol[:,0]
y2 = sol[:,1]
```

```
fig = plt.figure(facecolor='white')
ax = fig.add_subplot(111)
ax.set(title='Решение ОДУ')
ax.set_xlabel('$t$')
ax.set_ylabel('$f$')
ax.plot(ti, y1, color='c')
ax.plot(ti, y2, color='b')
lgnd = ax.legend(['y1', 'y2'], loc='upper right', shadow=True)
lgnd.get_frame().set_facecolor('#ffb19a')
plt.grid(True)
plt.show()
```
